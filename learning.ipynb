```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Step 1: Create a directed graph (since relationships have direction)
G = nx.DiGraph()

# Step 2: Add entities (nodes) and relationships (edges)
relations = [
    ("Elon Musk", "founder_of", "SpaceX"),
    ("SpaceX", "launches", "Falcon 9"),
    ("SpaceX", "owns", "Starlink"),
    ("Elon Musk", "CEO_of", "Tesla"),
    ("Tesla", "produces", "Electric Cars"),
    ("Falcon 9", "used_for", "Satellite Launch")
]

# Step 3: Add edges to the graph
for src, relation, target in relations:
    G.add_edge(src, target, label=relation)

# Step 4: Draw the graph
pos = nx.spring_layout(G, seed=64)  # positioning for clarity

plt.figure(figsize=(10, 6))
nx.draw(
    G, pos, with_labels=True, node_color="skyblue",
    node_size=2500, font_size=10, font_weight="bold", arrows=True
)

# Step 5: Add edge labels (the relationships)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")

plt.title("Mini Knowledge Graph Example", fontsize=14)
plt.show()
```
